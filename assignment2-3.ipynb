```python
import openai
import requests

client = openai.OpenAI()
messages = []
```

```python
def get_popular_movies(): #- /moviesм—җм„ң мқёкё° мҳҒнҷ”лҘј к°Җм ёмҳөлӢҲлӢӨ.
    response = requests.get("https://nomad-movies.nomadcoders.workers.dev/movies")
    return response.json()

def get_movie_details(id): # - /movies/:idм—җм„ң мҳҒнҷ” м •ліҙлҘј к°Җм ёмҳөлӢҲлӢӨ.
    response = requests.get("https://nomad-movies.nomadcoders.workers.dev/movies/{id}")
    return response.json()
    
def get_movie_credits(id): # - /movies/:id/creditsм—җм„ң м¶ңм—°м§„ л°Ҹ м ңмһ‘м§„мқ„ к°Җм ёмҳөлӢҲлӢӨ.
    
    response = requests.get("https://nomad-movies.nomadcoders.workers.dev/movies/{id}/credits")
    return response.json()


PROMPT = """
I have the following functions in my system.

'get_popular_movies' : мқёкё° мҳҒнҷ”лҘј к°Җм ёмҳөлӢҲлӢӨ. argumentлҠ” м—ҶмҠөлӢҲлӢӨ. dataлҠ” мҳҒнҷ” лҰ¬мҠӨнҠёмһ…лӢҲлӢӨ. к°Ғ мҳҒнҷ”лҠ” id, title, release_date, vote_average, overviewлҘј нҸ¬н•Ён•©лӢҲлӢӨ.

'get_movie_details' : мҳҒнҷ”лҰ¬мҠӨнҠёмқҳ idлҘј argumentлЎң л°ӣм•„ мҳҒнҷ” м •ліҙлҘј к°Җм ёмҳөлӢҲлӢӨ. argumentлҠ” мҳҒнҷ”мқҳ idмһ…лӢҲлӢӨ.
    
'get_movie_credits' : мҳҒнҷ”лҰ¬мҠӨнҠёмқҳ idлҘј argumentлЎң л°ӣм•„ м¶ңм—°м§„ л°Ҹ м ңмһ‘м§„ м •ліҙлҘј к°Җм ёмҳөлӢҲлӢӨ. argumentлҠ” мҳҒнҷ”мқҳ idмһ…лӢҲлӢӨ.


к°Ғ н•ЁмҲҳлҠ” APIм—җм„ң лҚ°мқҙн„°лҘј к°Җм ёмҷҖм„ң JSON нҳ•мӢқмңјлЎң л°ҳнҷҳн•©лӢҲлӢӨ.

Please answer with the name of function and its arguments if you want to use any of the above functions to answer the user's question. 

Please get the id of movies from the 'get_popular_movies' function if the question is about a specific movie.

And use the 'get_movie_details' and 'get_movie_credits' functions to get more information about the movie if needed.

If you don't need to use any function, just answer the question directly.
"""


messages.append({"role": "system", "content": PROMPT})
```

```python
def call_ai():

    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=messages,
    )

    message = response.choices[0].message.content

    messages.append({"role": "assistant", "content": message})

    print(f"AI: {message}")
    
```

```python
while True:
    message = input("Send a message to the LLM: ")
    if message == "quit" or message == "q" :
        break
    else:
        messages.append({"role": "user", "content": message})
        print(f"User: {message}")
        call_ai()
```
